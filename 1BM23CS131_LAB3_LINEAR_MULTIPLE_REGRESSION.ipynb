```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv('housing_area_price.csv')
df
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')

new_df = df.drop('price',axis='columns')
new_df

price = df.price
price


reg = linear_model.LinearRegression()
reg.fit(new_df,price)

reg.predict([[3300]])

reg.coef_

reg.intercept_

reg.predict([[5000]])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("canada_per_capita_income.csv")

# Check data
print(df.head())

# Independent and dependent variables
X = df[['year']]          # feature
y = df['per capita income (US$)']   # target

# Create model
model = LinearRegression()
model.fit(X, y)

# Predict income for year 2020
prediction_2020 = model.predict([[2020]])

print("Predicted Per Capita Income for 2020:", prediction_2020[0])

# Optional: Plot graph
plt.scatter(df.year, df['per capita income (US$)'])
plt.plot(df.year, model.predict(X))
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("hiring.csv")

print(df.head())

# ---------------------------
# Convert word numbers to digits
# ---------------------------

word_to_num = {
    'zero':0,
    'one':1,
    'two':2,
    'three':3,
    'four':4,
    'five':5,
    'six':6,
    'seven':7,
    'eight':8,
    'nine':9,
    'ten':10
}

df['experience'] = df['experience'].map(word_to_num)

# Fill missing experience with 0
df['experience'] = df['experience'].fillna(0)

# Fill missing test scores with median
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(
    df['test_score(out of 10)'].median()
)

# ---------------------------
# Define X and Y
# ---------------------------

X = df[['experience',
        'test_score(out of 10)',
        'interview_score(out of 10)']]

y = df['salary($)']

# ---------------------------
# Train Model
# ---------------------------

model = LinearRegression()
model.fit(X, y)

# ---------------------------
# Predictions
# ---------------------------

# 2 yr exp, 9 test, 6 interview
salary1 = model.predict([[2, 9, 6]])

# 12 yr exp, 10 test, 10 interview
salary2 = model.predict([[12, 10, 10]])

print("Predicted Salary 1:", salary1[0])
print("Predicted Salary 2:", salary2[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("1000_Companies.csv")

print(df.head())

# -----------------------
# Convert Categorical 'State' to Dummy Variables
# -----------------------

df = pd.get_dummies(df, drop_first=True)

# -----------------------
# Define X and Y
# -----------------------

X = df.drop('Profit', axis=1)
y = df['Profit']

# -----------------------
# Train Model
# -----------------------

model = LinearRegression()
model.fit(X, y)

# -----------------------
# Predict Profit
# -----------------------

# IMPORTANT:
# After get_dummies(drop_first=True)
# If states are: California, Florida, New York
# Then only two columns remain (example):
# State_Florida, State_New York

# For Florida:
# State_Florida = 1
# State_New York = 0

profit_prediction = model.predict([[91694.48, 515841.3, 11931.24, 1, 0]])

print("Predicted Profit:", profit_prediction[0])
```
